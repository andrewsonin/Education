# Семинар №2. Точечная оценка параметров

Пусть, задано семейство функций распределения $\{F_\xi(x,\theta),\theta\in\Theta\}$, где $\Theta$ &mdash; множество возможных значений параметра $\theta$.

Пусть, $X_1,..,X_n$ &mdash; реализация выборки **независимых в совокупности одинаково распределённых** (i.i.d.) случайных величин $\xi_1,..,\xi_n\sim\xi$, элементы которой имеют распределение $F_\xi(x,\theta_0)$ при некотором неизвестном значении $\theta_0\in\Theta$.

Задача состоит в том, чтобы оценить (восстановить) $\theta_0$ по выборке $X_1,..,X_n$ наиболее точно.

---
## Несмещённость
### Пример №1
Пусть, $\xi\sim\{\xi_i\}_{i=1}^n$ &mdash; i.i.d. случайные величины, распределённые непрерывно на отрезке $[0,\theta]$, то есть $\xi\sim\mathrm{U}[0,\theta]$

Придумать очевидные способы оценить $\theta$ по выборке.
1. $\hat{\theta}=\max(X_1,..,X_n)$
2. $\hat{\theta}=\large\frac{2}{n}\normalsize(X_1+...+X_n)$

---

### **<span style="color:blue">Определение №1</span>**

Оценка $\hat{\theta}(X_1,..,X_n)$ параметра $\theta$ называется **несмещённой**, если $\mathbb{E}\hat{\theta}(\xi_1,..,\xi_n)=\theta$ для всех $\theta\in\Theta$.

---

Матожидание берётся при условии, что все $\xi_i$ распределены с функцией распределения $F_\xi(x,\theta)$.

Условие "для всех $\theta\in\Theta$" существенно, поскольку тождественная оценка $\hat{\theta}(X_1,..,X_n)\equiv1$ для случайных величин из примера №1 идеальна при $\theta=1$, а при $\theta\neq1$ смещена.

## Состоятельность
### **<span style="color:blue">Определение №2</span>**

Оценка $\hat{\theta}(X_1,..,X_n)$ параметра $\theta$ называется **cостоятельной**, если $\forall\varepsilon>0$ справедливо $\lim\limits_{n\to+\infty}\mathbb{P}(|\hat{\theta}_n-\theta|>\varepsilon)=0$ для всех $\theta\in\Theta$.

---

### **<span style="color:red">Факт №1</span>**

Если смещение $b_n(\theta)=\mathbb{E}\hat{\theta}_n-\theta$ и дисперсия оценки $\mathbb{D}\hat{\theta}_n$ стремятся к нулю при увеличении размера выборки $n$, то оценка $\hat{\theta}_n$ состоятельна.

### **<span style="color:blue">Определение №3</span>**

Произвольная кусочно-непрерывная функция от выборки $(X_1,..,X_n)$ называется **статистикой**.

Статистика $T_n$ называется **асимптотически нормальной**, если $\exists a_n,b_n$, где $b_n>0$ такие, что $\large\frac{T_n-a_n}{b_n}\normalsize\xrightarrow[n\to\infty]{d}\zeta\sim N(0,1)$.

## Метод моментов

Пусть распределение элементов выборки зависит от $m$ неизвестных параметров $\theta_1,..,\theta_m$, и вектор $\boldsymbol{\theta}=(\theta_1,..,\theta_m)$ принадлежит некоторой области $\Theta\subset\mathbb{R}^n$. Пусть, $\mathbb{E}|\xi_\boldsymbol\theta|^m<\infty$ для всех $\boldsymbol\theta\in\Theta$. Тогда существуют все моменты предыдущих порядков, то есть $\forall k\in\{1,..,m\}\to\mathbb{E}\xi_\boldsymbol\theta^k<\infty$.

Можно записать систему уравнений $\mathbb{E}\xi_\boldsymbol\theta^k=A_k$, где $A_k=\large\frac{1}{n}\normalsize\sum\limits_{i}X_i^k$ &mdash; выборочные (эмпирические) моменты.

В левой части уравнений мы имеем теоретические значения моментов. Они являются функциями $m$ переменных $\theta_1,..,\theta_m$.

Справа записываются численные значения, посчитанные по результатам эксперимента.

Получается $m$ уравнений с $m$ неизвестными. Компоненты решения этого уравнения $\hat{\theta}_1,..,\hat{\theta}_m$ называются оценками метода моментов.

Есть определённый ряд требований к теоретическим значениям моментов, чтобы данный метод можно было применять:
- Отображение, задаваемое левыми частями уравнения, должно быть биективно: $\Theta\xrightarrow{\mathrm{ID}} P\subset\mathbb{R}^m$, где $P$ &mdash; некоторая область.
- Обратное отображение должно быть непрерывно.
- Тогда решение данной системы существует, единственно и непрерывно относительно правой части системы.

---

### **<span style="color:red">Факт №2</span>**

Можно доказать, что оценки, получаемые подобным методом **состоятельны**, а также, при ещё одном дополнительном условии\*, **асимптотически нормальны**.

<sup>*обратное отображение должно быть гладким</sup>

Пусть, наша выборка является реализацией i.i.d. случайных величин, распределённых, как $\xi\sim Norm(\mu,\sigma^2)$.

Посчитаем первый момент нормального распределения. По-определению, это матожидание
$\mathbb{E}\xi=\mu$.

Второй момент это $\mathbb{E}\xi^2$. Учитывая, что дисперсия $\mathbb{D}\xi=\mathbb{E}\xi^2-(\mathbb{E}\xi)^2$, получаем $\mathbb{E}\xi^2=\mathbb{D}\xi+(\mathbb{E}\xi)^2$. Зная, что дисперсия нормального распределения равна $\sigma^2$, получаем

$\mathbb{E}\xi^2=\sigma^2+\mu^2$.

В итоге это нам даёт следующую систему уравнений:

$\hat\mu=\large\frac{1}{n}\normalsize\sum\limits_{i=1}^nX_i$,

$\hat\sigma^2+{\hat\mu}^2=\large\frac{1}{n}\normalsize\sum\limits_{i=1}^nX_i^2$.

Откуда:

$\hat\mu=\large\frac{1}{n}\normalsize\sum\limits_{i=1}^nX_i$,

$\hat\sigma^2=\large\frac{1}{n}\normalsize\sum\limits_{i=1}^nX_i^2-{\hat\mu}^2=\large\frac{1}{n}\normalsize\sum\limits_{i=1}^nX_i^2-(\large\frac{1}{n}\normalsize\sum\limits_{i=1}^nX_i)^2$.

## Метод максимального правдоподобия

Пусть, $X_1,..,X_n$ &mdash; реализация выборки i.i.d. случайных величин $\xi_1,..,\xi_n\sim\xi$, элементы которой имеют либо дискретную, либо абсолютно непрерывную плотность вероятности распределения $p_\xi(x,\boldsymbol\theta)$, где $\boldsymbol\theta=(\theta_1,..,\theta_n)$ неизвестен.

Введём следующую функцию (называемой функцией правдоподобия):

$L(X, \boldsymbol\theta)=\prod\limits_{i=1}^np_{\xi}(X_i, \boldsymbol\theta)$,

где

- $X=(X_1,..,X_n)$ &mdash; выборка,
- $p_{\xi}(X_i, \boldsymbol\theta)$ &mdash; плотность вероятности случайной величины $\xi$ с подставленной в неё значением элемента выборки $X_i$.

Метод максимального правдоподобия находит $\boldsymbol\theta$ максимизацией функции правдоподобия: $\boldsymbol{\hat{\theta}}=\underset{\boldsymbol\theta}{\mathrm{argmax}}(L(X, \boldsymbol\theta))$

Обычно проще искать максимум для логарифма функции правдоподобия.

При выполнении некоторых условий, называемых условиями **регулярности**, оценки подобным методом получаются состоятельными и асимпотически нормальными.

## Функция риска

Функция риска вводится как мера "разброса" оценки параметра от его реального значения.

Функция $\mathrm{MSE_\hat{\boldsymbol\theta}}(\boldsymbol\theta)=\mathbb{E}(\hat{\boldsymbol\theta}-\boldsymbol\theta)^2$ называется среднеквадратичным риском.

Это функция именно от неизвестного параметра (или набора параметров) $\boldsymbol\theta$, и разброс нашей оценки $\hat{\boldsymbol\theta}$ от истинного значения $\boldsymbol\theta$ для разных $\boldsymbol\theta$, вообще говоря, разный.
