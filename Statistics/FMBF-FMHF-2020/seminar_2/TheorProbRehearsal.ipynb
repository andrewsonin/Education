# Повторение теории вероятностей
---
## **<span style="color:blue">Определение №1</span>**

Вероятностным пространством называется тройка элементов $(\Omega,\mathbb{F},\mathbb{P})$, где
- $\Omega$ &mdash; произвольное непустое множество, называемое пространством элементарных исходов,
- $\mathbb{F}$ &mdash; борелевская (минимальная) сигма-алгебра на $\Omega$,
- $\mathbb{P}$ &mdash; вероятностная мера $\mathbb F\to[0,1]$ такая, что
  - $\mathbb P(\Omega)=1$,
  - $\mathbb P(\bigsqcup\limits_{i=1}^\infty A_i)=\sum\limits_{i=1}^\infty \mathbb P(A_i)$ для для любой последовательности $\{A_i\}_{i=1}^\infty$ непересекающихся подмножеств $\Omega$ из $\mathbb F$.

## **<span style="color:blue">Определение №1.1</span>**

Система подмножеств $\mathbb{F}$ множества $\Omega$ называется **алгеброй** на $\Omega$, если выполнены следующие свойства:
- $\Omega\in\mathbb{F}$
- $A\in\mathbb F\Rightarrow\overline A\in\mathbb F$
- $A,B\in\mathbb F\Rightarrow A\cap B\in\mathbb F, A\cup B\in\mathbb F$ (достаточно только одного)

$\mathbb F$ называется **$\sigma$-алгеброй**, если для любой последовательности подмножеств $\{A_i\}_{i=1}^\infty\in\mathbb F$ справедливо как $\bigcap\limits_{i=1}^\infty A_i\in\mathbb F$, так и $\bigcup\limits_{i=1}^\infty A_i\in\mathbb F$.

Второе условие избыточно, поскольку $\bigcup\limits_{i=1}^\infty A_i=\Omega\setminus\bigcap\limits_{i=1}^\infty(\Omega\setminus A_i)$.

---

Зачем нужно вводить такую конструкцию, как $\mathbb{F}$? Очень просто: зачастую нам требуется каким-то образом классифицировать элементарные исходы и вычислить вероятность того, что элементарный исход принадлежит к данному конкретному классу. Нестрогий смысл $\mathbb{F}$ таков: это множество подмножеств $\Omega$. Элемент $A\in\mathbb{F}$ называется событием.

Например, при подбрасывании игрального кубика может произойти 6 элементарных исходов. Допустим, по условиям игры, выпавшее число $\geq5$ означает выигрыш, остальное &mdash; проигрыш. "Выигрыш" является событием, содержащим 2 элементарных исхода: 5 и 6. Его вероятность равна сумме вероятностей этих исходов:

$\mathbb{P}(\mathrm{Win})=\mathbb{P}\{\omega|\omega\in\mathrm{Win}\}=\sum\limits_{\omega\in\mathrm{Win}}\mathbb{P}(\omega)=\mathbb{P}(5)+\mathbb{P}(6)=\large\frac{1}{3}.$

Зачастую в реальных задачах работать с $\omega\in\Omega$ крайне неудобно, и поэтому используют численные характеристики исходов &mdash; случайные величины.

---
## **<span style="color:blue">Определение №2</span>**
**Cлучайной величиной $\xi$** называется измеримая функция $\Omega\xrightarrow{\xi}\mathbb{R}$.

---

Например:
1. При приёме на работу наниматели оценивают кандидатов по определённым критериям.

   Факт приёма (или неприёма) является результатом их впечатления от собеседуемых.

   Конкретное субъективное впечатление можно рассматривать как элементарный исход (элемент $\Omega$). Позитивные, умеренные и негативные впечатления &mdash; как события, классифицирующие элементарные исходы. Факт приёма в данном случае является случайной величиной, принимающей значения 0 в случае отказа и 1 в случае успеха, и она зависит от субъективного впечатления.

   Факт приёма &mdash; величина **наблюдаемая**.

   Конкретное впечатление нанимателей (их совокупность эмоций) &mdash; зачастую **скрытая**.


2. Имеется идеальный газ в изолированном сосуде.

   Измеряется среднее давление $P(t)$, оказываемое газом за предыдущую миллисекунду на площадку сосуда в зависимости от времени.

   В классической физике это однозначная функция времени и координат и импульсов всех частиц в начальный момент времени. То есть
   
   $P(t)=f(t,\overrightarrow{p_1},..\overrightarrow{p_n},\overrightarrow{r_1},..\overrightarrow{r_n})$, 
   
   где $\overrightarrow{r_i}, \overrightarrow{p_i}$ &mdash; радиус-вектор и импульс i-й частицы в начальный момент времени.
   
   Если в сосуде всего $n$ частиц, то $P(t)$ зависит от $6n+1$ переменных, где $n$ имеет характерный масштаб $n\sim10^{23}$.
   
   Каждая конкретная конфигурация $\{\overrightarrow{p_1},..\overrightarrow{p_n},\overrightarrow{r_1},..\overrightarrow{r_n}\}$ &mdash; это элементарный исход.

   Давление в конкретный момент времени на конкретную площадку сосуда &mdash; это случайная величина, определённая на множестве конфигураций радиус-веторов и импульсов всех частиц в начальный момент времени.
   
   Конфигурацию измерить **невозможно**, а давление &mdash; запросто!

   Этот пример также показывают, насколько сложными могут быть зависимости случайных величин от элементарных исходов.

## **<span style="color:red">Факт №1</span>**

Кусочно-непрерывная* функция случайной величины так же является случайной величиной.

<sup>*в общем случае: борелевская</sup>

## **<span style="color:blue">Определение №3</span>**

Два события $A,B\in\mathbb{F}$ называются независимыми, если $\mathbb{P}(A\cap B)=\mathbb{P}(A)\mathbb{P}(B)$

---
Это означает, что вероятность совместного осуществления как события $A$, так и события $B$ равна произведению вероятности их осуществления по-отдельности.

Это значит, что факты осуществления или не осуществления $A$ и $B$ не влияют друг на друга.

Несмотря на кажущуюся интуитивность этого понятия, оно и близко таковым не является, и в задачах теории вероятности независимость проверяется **только по определению**, то есть проверкой выполнения тождества выше.

---
## **<span style="color:red">Факт №2</span>**

Пусть, $\xi,\zeta$ &mdash; независимые случайные величины.

Тогда для любых кусочно-непрерывных* функций $f,g:\mathbb{R}\to\mathbb{R}$ случайные величины $f(\xi),g(\zeta)$ независимы.

Как следствие, $f(\xi)$ независима с $f(\zeta)$.

<sup>*в общем случае: для борелевских</sup>

---
## **<span style="color:blue">Определение №4</span>**

События $\{A_i\}_{i=1}^n$, где $n\in\mathbb{N}\cup\{+\infty\}$, называются **независимыми в совокупности**, если для любого конечного набора уникальных событий $\{A_{i_j}\}_{j=1}^k\subset\{A\}_i^n$, где $k\in\mathbb N,k\leq n$, верно $\mathbb{P}(\bigcap\limits_{j=1}^kA_{i_j})=\prod\limits_{j=1}^k\mathbb{P}(A_{i_j})$.

---
Из простой попарной независимости событий, вообще говоря, не следует независимость в совокупности.

Пример: Пусть брошены три уравновешенные монеты. Определим события следующим образом:
- $A_1$: монеты 1 и 2 упали одной и той же стороной;
- $A_2$: монеты 2 и 3 упали одной и той же стороной;
- $A_3$: монеты 1 и 3 упали одной и той же стороной;

Легко проверить, что любые два события из этого набора независимы. Все же три в совокупности зависимы, ибо зная, например, что события $A_1$ и $A_2$ произошли, мы знаем точно, что $A_3$ также произошло.

---
## **<span style="color:red">Факт №3</span>**

Пусть, $\{\xi_i\}_{i=1}^n$ &mdash; независимые в совокупности случайные величины.

Тогда для любых кусочно-непрерывных* функций $\{f_i\}_{i=1}^n:\mathbb{R}\to\mathbb{R}$ случайные величины $\{f_i(\xi_i)\}_{i=1}^n$ независимы в совокупности.

<sup>*в общем случае: для борелевских</sup>

## **<span style="color:blue">Определение №5</span>**

Пусть, $A,B\in\mathbb F$. Тогда в тождестве $\mathbb P(A\cap B)=\mathbb P(A|B)\mathbb P(B)$ член $P(A|B)$ называется условной вероятностью события $A$ при условии осуществления события $B$.

---

Смысл условной вероятности таков: есть два события $A$ и $B$. Они имеют какие-то вероятности, и мы хотим узнать, как изменится вероятность осуществления события $A$, если известно, что событие $B$ осуществилось. Эта "новая" вероятность события $A$ и есть условная вероятность $\mathbb P(A|B)$.

Очевидно, что для независимых событий $A$ и $B$ условная вероятность $A$ при условии $B$ никак не отличается от априорной $\mathbb P(A|B)=\mathbb P(A)$.

Как уже оговаривалось, независимость в терминах теории вероятности &mdash; вещь зачастую крайне неинтуитивная. В качестве примера можно привести следующий [парадокс](https://ru.wikipedia.org/wiki/Парадокс_Монти_Холла).

---

## **<span style="color:red">Факт №4. Формула Байеса</span>**

Пусть, $A,B\in\mathbb F$. Тогда

$\mathbb P(A|B)\mathbb P(B)=\mathbb P(B|A)\mathbb P(A)$.

---

## **<span style="color:red">Факт №5. Формула полной вероятности</span>**

Пусть, $\{A_i\}_{i=1}^n\subset\mathbb F$ &mdash; набор событий таких, что:
- $\mathbb P(A_i)>0$,
- $\forall i\neq j\to A_i\cap A_j=\emptyset$,
- $\bigsqcup\limits_{i=1}^n A_i=\Omega$,

а $B\in\mathbb F$ &mdash; какое-то событие. Тогда

$\mathbb P(B)=\sum\limits_{i=1}^n\mathbb P(B|A_i)\mathbb P(A_i)$.

## **<span style="color:blue">Определение №6</span>**

Функцией распределения $F_\xi(x)$ случайной величины $\xi$ называется $F_\xi(x)=\mathbb{P}(\xi^{-1}\{(-\infty;x]\})$, где $\xi^{-1}(A)$ означает полный прообраз множества $A\in\mathbb{B}(\mathbb{R})$ в множестве событий $\mathbb{F}$.

$\mathbb{B}(\mathbb{R})$ означает борелевскую сигма-алгебру на числовой прямой (все числа, интервалы, полуинтервалы, отрезки и их счётные объединения).

---

Менее формальный смысл функции распределения таков: $F_\xi(x)$ означает вероятность осуществления события $\{\omega|\xi(\omega)\leq x\}$ или, для краткости, $\mathbb{P}(\xi\leq x)$.

---
## **<span style="color:blue">Определение №7</span>**

Распределение случайной величины $\xi\in\mathbb R$ называется **абсолютно непрерывным**, если существует неотрицательная функция $p_\xi:\mathbb{R}\xrightarrow{p_\xi}\mathbb{R}_+$ такая, что $\forall x\in\mathbb{R}$ справедливо $F_\xi(x)=\large\int\limits_{-\infty}^x\normalsize p_\xi(t)\mathrm dt$. Функция $p_\xi$ называется **плотностью распределения вероятностей случайной величины**.

---

$p_\xi(x)=\large\frac{\mathrm dF_\xi(x)}{\mathrm dx}$ (в общем случае это производная слева).

---

## **<span style="color:blue">Определение №8</span>**

Матожидание случайной величины $\xi$

- Абсолютно непрерывный случай: $\mathbb{E}\xi=\large\int\limits_{-\infty}^{+\infty}\normalsize xp_\xi(x)\mathrm dx$
- Дискретный: $\mathbb{E}\xi=\sum\limits_{x=-\infty}^{+\infty} \mathbb{P}(\xi=x)*x$

Дисперсия случайной величины $\mathbb{D}\xi=\mathbb{E}(\xi-\mathbb{E}\xi)^2$

---

## **<span style="color:red">Факт №6</span>**

Для $\forall a_i\in\mathbb R$ и для любых случайных величин $\xi_i$ справедливо

$\mathbb{E}(a_1\xi_1+...+a_n\xi_n)=a_1\mathbb{E}\xi_1+...+a_n\mathbb{E}\xi_n$

---

## **<span style="color:blue">Определение №9</span>**

Две случайные величины $\xi,\zeta$ называются некоррелированными, если $\mathbb{E}(\xi\zeta)=\mathbb{E}\xi\mathbb{E}\zeta$

---
## **<span style="color:red">Факт №7</span>**

Из незавимости двух случайных величин следует некоррелированность, из некоррелированности независимость &mdash; нет.

---
## **<span style="color:red">Факт №8</span>**

Пусть ${\xi_1,..,\xi_n}$ &mdash; попарно некоррелированные случайные величины, $a_i\in\mathbb R$. Тогда

$\mathbb{D}(a_1\xi_1+...+a_n\xi_n)=a_1^2\mathbb{D}\xi_1+...+a_n^2\mathbb{D}\xi_n$
